```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import fetch_lfw_people
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.decomposition import PCA
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import fetch_lfw_people

def load_data():
    faces = fetch_lfw_people(min_faces_per_person=60)
    print('data loaded')
    print(faces.target_names)
    print(faces.images.shape)
    return faces
```

```python
faces=load_data()
```

```python
n_samples, h, w = faces.images.shape
```

```python
x=faces.data
n_features = x.shape[1]
n_features
```

```python
y = faces.target
target_names = faces.target_names
n_classes = target_names.shape[0]
n_classes
```

```python
print("Total dataset size:")
print("n_samples: %d" % n_samples)
print("n_features: %d" % n_features)
print("n_classes: %d" % n_classes)
```

```python
from sklearn.svm import SVC
from sklearn.decomposition import PCA as RandomizedPCA
from sklearn.pipeline import make_pipeline

pca = RandomizedPCA(n_components=150, whiten=True, random_state=42)
svc = SVC(kernel='rbf', class_weight='balanced')
model = make_pipeline(pca, svc)
```

#### 1. Split the data into a training and testing set.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=12)
```

#### 2.Use a grid search cross-validation to explore combinations of parameters to determine the best model:

C: margin hardness E.g. [1, 5, 10, 50]
gama: radial basis function kernel E.g. [0.0001, 0.0005, 0.001, 0.005]

```python
param_grid = {
    'pca__n_components': [5, 15, 30, 45, 64],
    'svc__C': [1,5,10,50],
    'svc__gamma' :['scale', 'auto']
}
```

```python
clf = GridSearchCV(model, param_grid, n_jobs=-1)
best=clf.fit(x_train, y_train)
```

```python
y_pred = best.predict(x_test)
```

#### precision

```python
from sklearn.metrics import precision_score, recall_score,f1_score 
precision_score(y_test, y_pred, average='macro')
```

#### recall

```python
recall_score(y_test, y_pred, average='macro')
```

#### f1-score

```python
f1_score(y_test, y_pred, average='macro')
```

#### support

```python
from sklearn.metrics import precision_recall_fscore_support
precision_recall_fscore_support(y_test, y_pred, average=None)[3]
```

#### 3. Draw a 4x6 subplots of images using names as label with color black for correct instances and red for incorrect instances.

```python
def plot_gallery(images, actual, predicted, h, w, n_row=4, n_col=6):
    
    plt.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    plt.subplots_adjust(bottom=0, left=.01, right=.99, top=.90, hspace=.35)
    for i in range(n_row * n_col):
        plt.subplot(n_row, n_col, i + 1)
        plt.imshow(images[i].reshape((h, w)), cmap=plt.cm.gray)
        if actual[i]==predicted[i]: 
            plt.title(predicted[i], size=12, fontdict={'color': 'black'})
        else:
            plt.title(predicted[i], size=12, fontdict={'color': 'red'})
        plt.xticks(())
        plt.yticks(())


def title(y_pred, y_test, target_names, i):
    pred_name = target_names[y_pred[i]].rsplit(' ', 1)[-1]
    true_name = target_names[y_test[i]].rsplit(' ', 1)[-1]
    return pred_name, true_name

prediction_titles = [title(y_pred, y_test, target_names, i)
                     for i in range(y_pred.shape[0])]

actual=[]
predicted=[]
for element in prediction_titles:
    predicted.append(element[0])
    actual.append(element[1])
    
plot_gallery(x_test, actual, predicted, h, w)
```

#### 4. Draw a confusion matrix between features in a heatmap with X-axis of 'Actual' and Y-axis of 'Predicted'.

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8,6))
ax= plt.subplot()
sns.heatmap(cm.T, annot=True, fmt='g', ax=ax)
ax.xaxis.set_ticklabels(target_names, rotation=90)
ax.yaxis.set_ticklabels(target_names, rotation=0)
ax.set_ylabel('Predicted')
ax.set_xlabel('Actual');
```
